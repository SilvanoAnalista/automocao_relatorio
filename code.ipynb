```python
#PASSO 4 _ CALCULAR OS INDICADORES

#IMPORTAR A BASE DE DADOS
#SEPARADOR "sep"
tab=pd.read_csv(r'C:\Users\danir\Downloads\Compras.csv',sep=";")
display(tab)

#CALCULO DOS INDICADORES

#TOTAL GASTO 
total=tab["ValorFinal"].sum()

#QUATIDADE
quantidade=tab["Quantidade"].sum()

#PREÇO MEDIO
preco_medio=total/quantidade
```

```python
#EXIBIR

print("Preco Medio - ",preco_medio)
print("Valor Final - ",total)
print("Quantidade  - ",quantidade)
```
